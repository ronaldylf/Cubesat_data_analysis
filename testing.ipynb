```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    'Category': ['Category A', 'Category B', 'Category C'],
    'Expected': [50, 30, 40],
    'Real': [45, 35, 42]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Set the 'Category' column as the index (optional)
df.set_index('Category', inplace=True)

# Plotting the bar chart
ax = df.plot(kind='bar', color=['blue', 'orange'], width=0.8, figsize=(10, 6))

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Expected vs Real Data')

# Adding legend
ax.legend(['Expected', 'Real'])

# Show the plot
plt.show()
```
